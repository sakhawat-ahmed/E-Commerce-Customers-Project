```python
# Define features and target variable
X = df[['Avg. Session Length', 'Time on App', 'Time on Website', 'Length of Membership']]
y = df['Yearly Amount Spent']
```

```python
# Using model_selection.train_test_split from sklearn to split the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# Import LinearRegression from sklearn.linear_model
from sklearn.linear_model import LinearRegression
```

```python
# Creating an instance of a LinearRegression() model named lm
lm = LinearRegression()
```

```python
# Train/fit lm on the training data
lm.fit(X_train, y_train)
```

```python
# Print out the coefficients of the model
print('Coefficients: \n', lm.coef_)
```

```python
# Using lm.predict() to predict off the X_test set of the data
predictions = lm.predict(X_test)
```

```python
# Creating a scatterplot of the real test values versus the predicted values
plt.scatter(y_test, predictions)
plt.xlabel('Y Test (True Values)')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values')
plt.show()
```

```python
# Calculating the Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
# Calculate the explained variance score (R^2)
from sklearn.metrics import r2_score
print('R-squared:', r2_score(y_test, predictions))
```

```python
# Plotting a histogram of the residuals and checking for normal distribution
residuals = y_test - predictions

sns.histplot(residuals, bins=50, kde=True)
plt.xlabel('Residuals')
plt.title('Distribution of Residuals')
plt.show()
```

```python
# Additional normality check with Q-Q plot
from scipy import stats

stats.probplot(residuals, dist="norm", plot=plt)
plt.title('Q-Q Plot of Residuals')
plt.show()
```

```python
# Check summary statistics of residuals
print('Residuals Summary:')
print('Mean:', residuals.mean())
print('Standard Deviation:', residuals.std())
```
